```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load new CSV file

raw_csv_data=pd.read_csv(r"C:\Users\LIM TZE TA\VSCodeProjects\ML_Assignment\Classification\online_shoppers_intention_cleaned.csv")
df=raw_csv_data.copy()


# Set general style
sns.set(style="whitegrid", palette="pastel")

# Create a 2x2 grid for better readability
fig, axes = plt.subplots(2, 2, figsize=(14, 8))

# 'Administrative' distribution
sns.histplot(df['Administrative'], kde=True, bins=40, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Administrative Page Visits')
axes[0, 0].set_xlabel('Administrative')

# 'Administrative_Duration' distribution
sns.histplot(df['Administrative_Duration'], kde=True, bins=40, ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Duration of Administrative Page Visits')
axes[0, 1].set_xlabel('Administrative_Duration')

# 'Informational' distribution
sns.histplot(df['Informational'], kde=True, bins=40, ax=axes[1, 0])
axes[1, 0].set_title('Distribution of Informational Page Visits')
axes[1, 0].set_xlabel('Informational')

# 'ProductRelated' distribution
sns.histplot(df['ProductRelated'], kde=True, bins=40, ax=axes[1, 1])
axes[1, 1].set_title('Distribution of Product-Related Page Visits')
axes[1, 1].set_xlabel('ProductRelated')

# Improve layout and spacing
plt.tight_layout()
plt.show()

```

```python
# Create the correlation matrix (numeric columns only)
corr = df.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap - Online Shoppers Intention Dataset")
plt.show()
```

```python
# Distribution graphs for each attribute (simplified)
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

df[numeric_cols].hist(bins=30, figsize=(15, 10), color='skyblue')
plt.suptitle("Distribution of Numeric Features")
plt.show()
```

```python
# Box Plot of the correlation between PageValues (strongest feature correlated to Revenue) and Revenue
plt.figure(figsize=(8,6))
sns.boxplot(x='Revenue', y='PageValues', data=df)
plt.title("PageValues vs Revenue")
plt.xlabel("Revenue (Did Purchase?)")
plt.ylabel("PageValues")
plt.show()
```

```python
import matplotlib.pyplot as plt
# Count the amt. of times each value appears in the column
counts = df['Revenue'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(    
    counts,
    labels=counts.index,
    autopct='%1.1f%%',
    shadow = True
)

plt.title('Revenue (Did Purchase?)')
plt.show()
```
