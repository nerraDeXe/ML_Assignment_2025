```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_selection import SelectKBest, chi2

# Load new CSV file

raw_csv_data=pd.read_csv(r"C:\Users\LIM TZE TA\VSCodeProjects\ML_Assignment\Classification\online_shoppers_intention_cleaned.csv")
df=raw_csv_data.copy()
```

```python
# Ensure target is numeric (0/1)
df['Revenue'] = df['Revenue'].astype(int)

# Compute correlations with the target
corr = df.corr(numeric_only=True)['Revenue'].sort_values(ascending=False)

plt.figure(figsize=(8,6))
sns.barplot(x=corr.index, y=corr.values)
plt.xticks(rotation=90)
plt.title("Correlation of Features with Target (Revenue)")
plt.show()
```

```python
# Adjust font size globally
plt.rcParams.update({'font.size': 8}) 

# Plot histograms
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_cols].hist(bins=30, figsize=(15, 10), color='skyblue')

plt.suptitle("Distribution of Numeric Features", fontsize=12)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # ensures title fits nicely
plt.show()
```

```python
# Create the correlation matrix (numeric columns only)
corr = df.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap - Online Shoppers Intention Dataset")
plt.show()
```

```python
# Box Plot of the correlation between PageValues (strongest feature correlated to Revenue) and Revenue
plt.figure(figsize=(8,6))
sns.boxplot(x='Revenue', y='PageValues', data=df)
plt.title("PageValues vs Revenue")
plt.xlabel("Revenue (Did Purchase?)")
plt.ylabel("PageValues")
plt.show()
```

```python
# Count the amt. of times each value appears in the column
counts = df['Revenue'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(    
    counts,
    labels=counts.index,
    autopct='%1.1f%%',
    shadow = True
)

plt.title('Revenue (Did Purchase?)')
plt.show()
```
