```python
import pandas as pd

df = pd.read_csv(r"C:\Users\LIM TZE TA\VSCodeProjects\ML_Assignment\Classification\online_shoppers_intention.csv")

# Display dataset in a table
df.copy()
```

```python
# Drop duplicated rows
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
# Describe dataset in terms of statistical terms
df.describe()
```

```python
# Double-check all unique values in the 'Month' column

unique_values = df['Month'].unique()
print(unique_values)

# Double-check all unique values in the 'Visitor' column

unique_values = df['VisitorType'].unique()
print(unique_values)
```

```python
# Map the months and change the values from object to int

month_map = {
    'Feb': 1, 'Mar': 2, 'May': 3, 'June': 4, 'Jul': 5, 
    'Aug': 6, 'Sep': 7, 'Oct': 8, 'Nov': 9, 'Dec': 10,
}

df['Month'] = df['Month'].map(month_map)

# Map the visitor type and change the values from object to int

visitor_map = {
    'Returning_Visitor': 1, 'New_Visitor': 2, 'Other': 3
}

df['VisitorType'] = df['VisitorType'].map(visitor_map)
```

```python
import numpy as np

# Round off some columns to less decimal places
df['ProductRelated_Duration'] = np.ceil(df['ProductRelated_Duration'] * 100) / 100
df['PageValues'] = np.ceil(df['PageValues'] * 10000) / 10000
df['Informational_Duration'] = np.ceil(df['Informational_Duration'] * 10000) / 10000
df['Administrative_Duration'] = np.ceil(df['Administrative_Duration'] * 10000) / 10000
df['ExitRates'] = np.ceil(df['ExitRates'] * 10000) / 10000
df['BounceRates'] = np.ceil(df['BounceRates'] * 10000) / 10000
```
